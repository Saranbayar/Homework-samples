- Make a feature correlation plot of the features 'sepal length', 'sepal width', and 'petal width'.
- Reduce the dimensionality of your features via PCA and via t-SNE downprojection. For t-SNE, use a perplexity value of 25.

```python
features = [r'sepal length', r'sepal width', r'petal width']
_ = sns.pairplot(iris_data, vars=features, hue=r'target', palette="deep")
```

```python
np.random.seed(seed=50)
iris_data_pca = u1.apply_pca(n_components=2, data=iris_data, target_column=r'target')
u1.plot_points_2d(data=iris_data_pca, target_column=r'target')
```

```python
np.random.seed(seed=50)
iris_data_tsne = u1.apply_tsne(n_components=2, data=iris_data, perplexity=25.0, target_column=r'target')
u1.plot_points_2d(data=iris_data_tsne, target_column=r'target')
```
